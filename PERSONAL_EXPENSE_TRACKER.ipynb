```python
import csv
import os
from datetime import datetime
from collections import defaultdict
```

```python
# --- Step 1 & 2: Project Setup & Expense Structure ---
FILE_NAME = 'expenses.csv'
HEADERS = ['date', 'amount', 'category']
```

```python
def init_file():
    """Creates the CSV file with headers if it doesn't exist."""
    if not os.path.exists(FILE_NAME):
        with open(FILE_NAME, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(HEADERS)
```

```python
# --- Step 3: Save to CSV ---
def add_expense():
    """Prompts the user for expense details and saves them to the CSV."""
    try:
        amount = float(input("💰 Enter amount: "))
    except ValueError:
        print("❌ Invalid amount. Please enter a number.")
        return

    category = input("🏷️  Enter category (e.g., Food, Travel, Rent): ").strip().title()
    date_str = input("📅 Enter date (YYYY-MM-DD) or press Enter for today: ").strip()

    if not date_str:
        date_str = datetime.today().strftime('%Y-%m-%d')

    # Using the dictionary structure from Step 2
    expense = {
        "date": date_str,
        "amount": amount,
        "category": category
    }

    with open(FILE_NAME, mode='a', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=HEADERS)
        writer.writerow(expense)

    print(f"✅ Expense of ₹{amount} added under '{category}' on {date_str}!\n")
```

```python
# --- Step 4 & 5: Read, Calculate Totals, and Add Summary ---
def view_summary():
    """Reads the CSV, calculates totals, and displays a summary."""
    if not os.path.exists(FILE_NAME):
        print("❌ No expenses found. Add some first!\n")
        return

    total_spend = 0.0
    category_totals = defaultdict(float)
    unique_dates = set()

    with open(FILE_NAME, mode='r') as file:
        reader = csv.DictReader(file)
        rows = list(reader)

        if not rows:
            print("❌ No expenses found. Add some first!\n")
            return

        for row in rows:
            amt = float(row['amount'])
            total_spend += amt
            category_totals[row['category']] += amt
            unique_dates.add(row['date'])
    # Data Analysis
    highest_category = max(category_totals, key=category_totals.get)
    days_recorded = len(unique_dates)
    avg_daily_spend = total_spend / days_recorded if days_recorded > 0 else 0

    # Display Summary
    print("\n📊 --- EXPENSE SUMMARY ---")
    print(f"💵 Total Spend: ₹{total_spend:.2f}")
    print(f"🔥 Highest Category: {highest_category} (₹{category_totals[highest_category]:.2f})")
    print(f"📈 Average Daily Spend: ₹{avg_daily_spend:.2f}/day (based on {days_recorded} days of data)")
    print("---------------------------\n")
```

```python
def export_report():
    """Exports a simple textual report (optional feature expansion)."""
    if not os.path.exists(FILE_NAME):
        print("❌ Nothing to export yet!\n")
        return

    report_name = f"report_{datetime.today().strftime('%Y%m%d')}.txt"
    with open(report_name, mode='w') as file:
        file.write("EXPENSE REPORT\n")
        file.write(f"Generated on: {datetime.today().strftime('%Y-%m-%d')}\n\n")

        # We can reuse the logic from summary here, or just dump the CSV contents
        with open(FILE_NAME, mode='r') as csv_file:
            reader = csv.reader(csv_file)
            for row in reader:
                file.write(" | ".join(row) + "\n")

    print(f"✅ Report exported successfully to {report_name}!\n")
```

```python
# --- Main Application Loop ---
def main_menu():
    init_file()
    while True:
        print("🚀 PERSONAL EXPENSE TRACKER")
        print("1️⃣ Add Expense")
        print("2️⃣ View Summary")
        print("3️⃣ Export Report")
        print("4️⃣ Exit")

        choice = input("👉 Choose an option (1-4): ").strip()

        if choice == '1':
            add_expense()
        elif choice == '2':
            view_summary()
        elif choice == '3':
            export_report()
        elif choice == '4':
            print("👋 Goodbye!")
            break
        else:
            print("❌ Invalid choice. Try again.\n")

if __name__ == "__main__":
    main_menu()
```
